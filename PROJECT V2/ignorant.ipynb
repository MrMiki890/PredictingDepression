```python
for i in range(0, len(df)):
    if df.iloc[i, -1] <= 13:
        df.iloc[i, -1] = 0
    elif 13 < df.iloc[i, -1]:
        df.iloc[i, -1] = 1
```

```python
for i in range(0, len(df)):
    if df.iloc[i, 10] <= 9:
        df.iloc[i, 10] = 0
    elif 9 < df.iloc[i, 10] <= 13:
        df.iloc[i, 10] = 1
    elif 13 < df.iloc[i, 10] <= 20:
        df.iloc[i, 10] = 2
    elif 20 < df.iloc[i, 10] <= 27:
        df.iloc[i, 10] = 3
    elif 27 < df.iloc[i, 10]:
        df.iloc[i, 10] = 4
```

```python
exp_clf = setup(
    data=df,
    target="depression",
    session_id=123,
    categorical_features=["urban", "gender", "religion", "orientation", "race", "voted", "married"],
    numeric_features=["age", "familysize", "extraversion", "conscientiousness", "agreeableness", "emotional_stability", "openness_to_experiences"],
    ordinal_features={
        "education": ["1", "2", "3", "4"]
    },
    fold_shuffle=True,
    use_gpu=True,
    feature_selection = True,
    feature_selection_threshold=0.999,
    data_split_stratify=True,
    normalize=True,
    transformation=True,
    log_data=True,
    log_experiment=True
)
```
