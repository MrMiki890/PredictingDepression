### For dataset version 1.02

New data profile with new dataset without Cholesterol, blood count, occupation, and prescription medications

```python
# New data profile with new dataset without 
# Cholesterol, blood count, occupation, and prescription medications

import pandas as pd
import pandas_profiling as pp

df = pd.read_csv("../datasets/dataset(08052022)-(1.02).csv")

# data profile
data_profile = pp.ProfileReport(
    df, 
    title="Data profiling before data preparation(1.01)",
    pool_size=12,
    dark_mode=True,
    correlations={
        "pearson": {"calculate": True},
        "spearman": {"calculate": True},
        "kendall": {"calculate": True},
        "phi_k": {"calculate": True},
        "cramers": {"calculate": True},
    },
    interactions = {
        "continuous":False
    },
    lazy=False
)

# save as html
data_profile.to_file('../others/data_profile_before_preparation(1.01).html')


# import streamlit as st
# from streamlit_pandas_profiling import st_profile_report
# # data_profile.to_notebook_iframe()
# # st.title("Pandas Profiling in Streamlit!")
# # st.write("profile.html", unsafe_allow_html = True)
# st_profile_report(data_profile)
```
